<h1 style=background-color:purple;color:white;text-align:center;font-family:'sans-serif'>IMPORT LIBRARIES</h1>

```python
import numpy as np 
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
x1 = np.linspace(0, 10, 100)


# create a plot figure
fig = plt.figure()

plt.plot(x1, np.sin(x1), '-')
plt.plot(x1, np.cos(x1), '--');
```

```python
# create a plot figure
plt.figure()


# create the first of two panels and set current axis
plt.subplot(2, 1, 1)   # (rows, columns, panel number)
plt.plot(x1, np.sin(x1))


# create the second of two panels and set current axis
plt.subplot(2, 1, 2)   # (rows, columns, panel number)
plt.plot(x1, np.cos(x1));
```

```python
# get current figure info

print(plt.gcf())
```

```python
# get current axis info

print(plt.gca())
```

```python
plt.plot([1, 2, 3, 4])
plt.ylabel('Numbers')
plt.show()
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16]) # plot() --> versatile command
plt.show()
```

```python
x = np.linspace(0, 2, 100)

plt.plot(x, x, label='linear')
plt.plot(x, x**2, label='quadratic')
plt.plot(x, x**3, label='cubic')

plt.xlabel('x label')
plt.ylabel('y label')

plt.title("Simple Plot")

plt.legend()

plt.show()
```

```python
plt.plot([2, 3, 4, 5], [4, 9, 16, 25], 'ro') # with red circles
plt.axis([0, 6, 0, 28])
plt.show()
```

```python
# evenly sampled time at 200ms intervals
t = np.arange(0., 5., 0.2)

# red dashes, blue squares and green triangles
plt.plot(t, t, 'g--', t, t**2, 'rs', t, t**3, 'b^')
plt.show()
```

```python
# First create a grid of plots
# ax will be an array of two Axes objects
f, ax = plt.subplots(2)

# Call plot() method on the appropriate object
ax[0].plot(x1, np.sin(x1), 'b-')
ax[1].plot(x1, np.cos(x1), 'b-');
```

```python
fig = plt.figure()  # creating a new axes

x2 = np.linspace(0, 5, 10)
y2 = x2 ** 2

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])

axes.plot(x2, y2, 'b--')

axes.set_xlabel('x2')
axes.set_ylabel('y2')
axes.set_title('title');
```

```python
fig = plt.figure() # fig is an instance of class which is plt.figure()
ax = plt.axes()
```

```python
# creating subplots

fig = plt.figure()

ax1 = fig.add_subplot(2,2,1)  # 2*2=4 plots
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)
```

```python
x3 = np.arange(0.0, 6.0, 0.01) 

plt.plot(x3, [xi**2 for xi in x3], 'p-') 

plt.show()
```

```python
x4 = range(1, 5) # plotting more than one plot on the same figure

plt.plot(x4, [xi*1.5 for xi in x4])

plt.plot(x4, [xi*3 for xi in x4])

plt.plot(x4, [xi/3.0 for xi in x4])

plt.show()
```

<h2 style=font-family:'sans-serif'>ERROR BAR CHART</h2>

```python
x9 = np.arange(0, 4, 0.2)

y9 = np.exp(-x9)

e1 = 0.1 * np.abs(np.random.randn(len(y9)))

plt.errorbar(x9, y9, yerr = e1, fmt = '.-')

plt.show();
```

<h2 style=font-family:'sans-serif'>STACKED BAR CHART</h2>

```python
A = [15., 30., 45., 22.]

B = [15., 25., 50., 20.]

z2 = range(4)

plt.bar(z2, A, color = 'b')
plt.bar(z2, B, color = 'r', bottom = A)

plt.show()
```

<h2 style=font-family:'sans-serif'>Adding a legend</h2>

```python
x15 = np.arange(1, 5)

fig, ax = plt.subplots()

ax.plot(x15, x15*1.5)
ax.plot(x15, x15*3.0)
ax.plot(x15, x15/3.0)

ax.legend(['Normal','Fast','Slow']);
```

```python
x15 = np.arange(1, 5)

fig, ax = plt.subplots()

ax.plot(x15, x15*1.5, label='Normal')
ax.plot(x15, x15*3.0, label='Fast')
ax.plot(x15, x15/3.0, label='Slow')

ax.legend();
```

```python
x15 = np.arange(1, 5)

fig, ax = plt.subplots()

ax.plot(x15, x15*1.5, label='Normal')
ax.plot(x15, x15*3.0, label='Fast')
ax.plot(x15, x15/3.0, label='Slow')

ax.legend(loc=1);
```
